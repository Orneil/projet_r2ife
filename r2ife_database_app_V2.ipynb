2 - Code source de l'application

*writefile app.py* permet ici d'écrire directement le code dans un fichier *app.py*. L'extension *.py* est effectivement necessaire au lancement de l'application streamlit

```python
%%writefile app.py

# Importer les bibliothèques nécessaires
import streamlit as st
import pandas as pd

# Chargement du fichier Excel depuis Google Drive (via le lien de partage)
google_drive_link = "https://docs.google.com/spreadsheets/d/1GgMH4_rNkMNAl-4FBpmbgWwvGVbE_1p2/edit?usp=drive_link"

# Extraction de l'ID du fichier depuis le lien
file_id = google_drive_link.split('/')[-2]

# Construction du lien de téléchargement direct
download_link = f'https://drive.google.com/uc?id={file_id}'

# Chargement du fichier Excel dans un DataFrame pandas
df = pd.read_excel(download_link)

# Affichage du titre de l'application
st.title("Base de données des ingénieurs")

# Ajout d'un sélecteur de colonnes pour la visualisation
columns_to_display = st.multiselect("Sélectionnez les colonnes à afficher", df.columns)

# Filtrer les données en fonction des critères sélectionnés
filtered_df = df[columns_to_display]

# Afficher les données filtrées
st.dataframe(filtered_df)

# Ajouter des options de filtrage supplémentaires
field_filter = st.sidebar.selectbox("Filtrer par Domaine de formation", df["Formation"].unique())
filtered_by_field = df[df["Formation"] == field_filter]

# Afficher les données filtrées par domaine de formation
st.write(f"Données filtrées pour le domaine de formation : {field_filter}")
st.dataframe(filtered_by_field)
```
